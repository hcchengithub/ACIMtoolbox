# 情境式冥想

 

除了每天跟著操練手冊運用當天的課程外，奇蹟課程的正文中也有許多可供我們平常遇到狀況時的提醒，茲提供參考如下。我們可以根據正文提供的某些狀況，加以運用於每天的冥想中或是協助我們安靜下來，重新看待問題。

## 當我面對誘惑而舉棋不定時

願我能就下列的問題反問自己，且做個決定，一切就會為他而完成：

我究竟渴望一個由我統治的世界，還是一個統治我的世界？

我究竟渴望一個令我強大有力的世界，或是舉目無援的世界？

我是否渴望一個沒有仇敵，並且不可能犯罪的世界？

我真的想要看見那因為是真理而被我否定的一切嗎？ 


## 當我看到弟兄的缺陷或過失時

我若想知道自己的真相，就必須這樣看待他。

天父，為禰完美的聖子，我感謝禰；我將在他的榮耀內看見自己的榮耀。 


## 當我自以為是時

世界最怕聽見的論調就是：

我不知道我是什麼，因此也不知道自己在做什麼，或身在何處，或是該如何看待世界或自己。

學到這一點，救恩就來臨了。


## 當我恨自己鐵不成鋼時

淨化純是上主的事，必是為你而設的。與其設法為祂準備自己 ，不妨這樣去想：

身為上主居所的我，是配得上祂的。

祂按照自己的意願，在我內建立了祂的居所。

我無須為祂作任何準備，只要我不去干擾祂的計劃，就能意識到自己無始以來即已準備妥善。

我無須在祂的計劃上畫蛇添足。

然而，若要接受它，我必須甘願不再以自己的計劃來取代祂的。


## 當我放不過對不起我的人時

向我的弟兄這樣說：

我把你視為自己的一部分，交託給聖靈。

我知道你終會得到解脫，只要我不利用你來囚禁我自己。

為了我的自由之故，我決心釋放你，因為我已認清了，我們將會一起獲得自由。


6.     當我為別人操心時

恐懼不論從那一個角度來看都是虛幻不實的，因為在創造的層次上，它既不存在，就根本不存在。你願讓你的信念接受這一考驗的程度有多深，你的知見便會得到多深的修正。在分辨真偽的過程，奇蹟是按照下面的邏輯進行的：

完美的愛驅逐了恐懼，

如果恐懼存在，

完美的愛就不存在。

然而：完美的愛是唯一的存在，

如果恐懼也存在，表示它已造出了一個根本不存在的境界。

只要相信這一點，你就自由了。


.     當我感到自己在付痛苦的代價時

記住這一點：

在這世界中，你無需苦難 的考驗，因為我已戰勝了世界。為此之故，你該歡樂才對。


8.     當我有意助人時

為了他人，也為了你自己的救治，你大有可為之處；面對需要你援助的境遇時，你該如此去想：

我在這兒，純粹是為了真正想要幫忙。

我在這兒，是代表派遣我的那一位。

我不用擔心該說什麼或做什麼，

派遣我來的那一位自會指點迷津。

祂希望我去的地方，我欣然前往，

因我知道祂與我同行。

只要我肯向祂學習救治之道，我便會得到救治。


9.     當我甩不掉內疚時

每當你好似被罪咎之苦所吸引，記住，只要你一向它屈服，就表示你決心放棄幸福，那你就無從獲得幸福了。因此，不妨本著天父及聖子之愛，堅定而溫和地向自己說：

凡是我經驗到的，我都願公諸於世。

我若是清白的，我就一無所懼。

我決心證明自己接受了救贖，不再加以排拒。

我願藉著顯示並分享我的清白，表達我的接納。

願我由天父那兒帶給上主之子平安。


10. 當我發現自己犯錯時

你的責任只是將你的思想帶回到當初犯錯的那個關卡，上，安心地將它託付給救贖。盡可能誠心地以下面的句子提醒自己，並記住，即使你的邀請只是輕描淡寫的一個暗示，聖靈仍會全力以赴地回應的：

我必定作了錯誤的決定，

因為我並沒有活在平安中。

是我自己作了這決定，

但我也能作出另一種決定。

我要作出另一種決定，因為我要活在平安中。

我沒有罪惡感，因為只要我給聖靈機會，

祂就會化解我錯誤的決定所帶來的一切後果。

我決定託付給祂，讓祂代我去選擇上主。 


 



11. 當我開始批判別人時

在任何境遇中，不論你是多麼想要判斷，並以此作為你反應的憑據，你都應謹記下列的想法，須臾不可忘失。把你的心思專注其上：

我不是孤獨的，

我不願意以過去的經驗來侵擾我的神聖貴賓。

我已邀請了祂，祂就在這兒。

只要不插手干預，其他我什麼都不必做。

12. 當我有逃避問題之念時

每當你忍不住想浪跡天涯，遠離光明時，不妨提醒一下自己你想要的究竟是什麼，並對自己說：

聖靈引導我到基督那裏去，

除了那兒，我還會想去那裏？

除了覺醒於祂以外，我還需要什麼？


13. 當我懷疑自己的價值時

當你懷疑自己的價值時，應說：

沒有我，連上主都無法圓滿。

當小我發言時，只要記住這一點，你就不會聽它那一套了。 


14. 當我心裡開始焦慮不安時

當你的平安受到任何威脅或騷擾時，便對自己說：

我不知道任何事物的意義，包括這件事在內。

因此，我也不知道如何回應。

此刻，我不願用過去的經驗作為引導我的光明。


15. 當我面臨與人衝突的危機時

當危機好似現前，神智比較清明的那一方，應該記得自己虧欠對方多深，深深感念對方的恩德，並且高興自己終有機會償還此債，把幸福帶給彼此。願他記住這一點，並說：

我自己渴望這神聖的一刻，

並願與我所愛的弟兄分享。

沒有他，我不可能擁有這一刻；

沒有我，他也不可能。

而此刻正是我們共享這神聖一刻的大好良機。

因此，我選擇這一刻，作為獻給聖靈的禮物，

使祂的祝福能降臨於我們身上，

讓我們一起活於平安中。


16. 當我當我面對令我憤怒或不安的外境時

你心裏有數，不論你看到罪的後果化身為何種形式，你只需要反問自己一下：

這可是我想要看到的？這是我所要的嗎？


17. 當我當我回顧自己一生而怨天尤人時

你只須如此說，但必須誠心誠意，而且毫不保留，因救恩的大能就在這些話中：

我應對自己所看見的一切負責。

我所經驗到的感受是出於自己的選擇，

我有意完成的目標也是出於自己的決定。

我所經歷到的一切，都是我自己招惹來的，

我所接受的也是自己祈求來的。

面對自己的遭遇，

不要再如此的無奈無助而自欺了。

你只須承認自己犯了錯誤，

從此就能不受其遺害所苦。


18. 當我面對不公平或不正義的事件時

當你在任何地方認出不義的景象，你只須如此說：

我若這樣看，等於否定了天父及聖子的神聖臨在。

我更願知道祂們的真相，而不願看到任何不義之事，因祂們臨在的光明已將它驅散了。


18. 當我用理性去分析或判斷別人時

面對看起來這樣不確定的意義，不要妄加判斷。記住，上天賜你的那位，其神聖的臨在，才是審判的終極之源。你應交託給祂去代你審判，並說：

把這個從我這兒拿走，幫我看管，並為我判斷吧！

願我不再把它視為罪與死亡的標誌，也不將它用於毀滅之途。

教導我如何使它不再成為平安的障礙，而讓禰為我發揮其用，為它的來臨鋪路。


19. 當我面對自己脆弱的致命傷時

當偶像再來引誘你時，記住這幾句話：

除了罪咎以外，偶像從未給過你其他的「禮物」。你為每個禮物都曾付出痛苦的代價，而且付出的不只你一個人而已。


20. 當過去的內疚或委屈開始浮現時

天父，寬恕我們心中的幻覺吧，幫助我們接受與禰的真正關係；這關係中沒有幻覺，幻覺也永遠無法侵入。我們的神聖本質原是來自於禰的。禰的神聖性既圓滿無缺，我們內心還需要什麼寬恕？遺忘與昏睡的心境，不過反映出我們不願憶起禰的寬恕與聖愛。願我們不再在誘惑中徘徊，因讓上主之子受誘，絕非禰的旨意。願我們只接受禰所賜的一切，內心也只接受禰所創造以及所愛的一切。阿們。


21. 當我不滿某位朋友時

如果基督在榮耀中向你現身，只要求你：

重新作個選擇吧！你究竟想要躋身於人間救主的行列，還是寧願拉著你的弟兄一起陷身於地獄？


22. 當我自哎自憐時

針對你還不時會把自己視為軟弱可憐的傾向，不妨養成習慣，學著用下面的話愉快地答覆：

我仍是上主所創造的我。上主之子不可能受任何苦。而我正是祂的聖子。 


 



23. 當我感到焦慮時

你若扭曲了實相，自會感受到焦慮、沮喪，終至驚惶失措，因為你企圖把自己變為非真之物。當你有這些感受時，不要企圖在自己之外尋求真理，真理只可能在你內。因此，你應如此說：

基督在我內，祂所在之處，上主必臨在，因為基督乃是祂的一部分。


24. 奇蹟學員每天的生活程式

只須想一想你究竟想要過什麼樣的日子，並告訴自己有一個方法能夠讓今天就變成那樣。接著，再試著把這一天變得像你所要的那樣。

（1）整個步驟是這樣開始的：

今天，我不靠自己作任何決定。

（2）從早到晚，任何時間，只要你想到這一點，就靜下來反省一會兒，再度提醒自己，你想要過的是什麼樣的日子，你想要的是什麼感受，你希望發生在你身上以及想要經歷的是什麼樣的事情，然後說：

只要我不自作主張，我就會得到這種日子。

 然而，難免有些時候，你已經先行判斷了。….

（3）再次提醒自己你想要過的是什麼日子，並認清眼前所發生的事與這種日子毫不相干。這樣，你便明白自己不只是已經提出問題，而且必然也已經按自己的方式答覆了。那麼，對自己說：

我沒有問題。我忘了該決定什麼。

（4）你若還不甘心接受，放不下自己的問題，那麼，下面這句話能夠幫你改變心意：

我至少能肯定一件事，就是：我並不喜歡自己目前的感受。

 （5）你既已肯定自己並不喜歡目前的感受，就不難繼續下去了：

因此，我希望我自己的那一套是錯的。

（6）這一點點智慧之芽，足以讓你更上一層樓。沒有人威脅或強迫你，你只是希望得到自己想要的東西。因此，你能誠心誠意地這樣說：

我願以另一種角度來看此事。

（7）最後一步只是確認自己不再拒絕接受援助了。這顯示出你開放的心胸，雖還不太肯定，但至少已甘心去面對真相了：

也許還有其他方式來看此事。

問一問，又有何妨？
